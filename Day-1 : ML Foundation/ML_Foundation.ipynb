Before we build ML models, let's brush up on:

```python
# Basic Arithmetic
a = 5
b = 10
print(f"The sum of {a} and {b} is: {a + b}")
print("The sum of {} and {} is: {}".format(a, b, a + b))
```

```python
# Variables and Data Types
name = "Alice"  # String
age = 30      # Integer
height = 5.7  # Float
is_student = False # Boolean
print(f"Name: {name}, Age: {age}, Height: {height}, Is student: {is_student}")
```

```python
# Dictionaries
student = {'name': 'Bob', 'major': 'CS', 'grade': 'A'}
print(f"Student dictionary: {student}")
print(f"Student's name: {student['name']}")
student['city'] = 'New York'
print(f"Student dictionary after adding city: {student}")
```

```python
# Loops(for)
for number in numbers:
    print(f"Square of {number} is: {number**2}")
```

```python
# Conditional Statements (if/else)
score = 75
if score >= 70:
    print("Pass")
else:
    print("Fail")
```

**4.1. NumPy** - NumPy is a general-purpose fundamental package for scientific computing with Python. It contains various features including these important ones:

- Array creation
- Reshaping array
- Flatten array

üîó Source: Official website of NumPy (https://numpy.org/)

```python
import numpy as np
```

```python
# Creating NumPy arrays
array_1d = np.array([1, 2, 3])
array_2d = np.array([[1, 2], [3, 4]])
print(f"1D NumPy array:\n{array_1d}")
print(f"2D NumPy array:\n{array_2d}", end='\n\n')

# Basic array operations
print(f"Sum of array_1d: {np.sum(array_1d)}")
print(f"Mean of array_2d: {np.mean(array_2d)}")
print(f"Element-wise multiplication:\n{array_1d * 2}")
```

```python
b1 = np.array([1,2,3,5,5,8])    #Declaring a NumPy Array
b2 = np.array([4,5,6,7,7,9])

print(b1+b2, end='\n\n')

print(b2 * 3, end='\n\n')
print("No. of dimensions: ", b1.ndim, end='\n\n')  # Rows in array, considered as a matrix.
# Printing shape of array
print("Shape of array: ", b1.shape, end='\n\n')  # Dimension

#reshaping an array
r_b1=b1.reshape(2,3)

print("Reshaped array: ", r_b1)
print("Shape of array: ", r_b1.shape, end='\n\n')

# Printing size (total number of elements) of array
print("Size of array: ", b1.size) # elements in a row or column elements.

# Printing the datatype of elements in array
print("Array stores elements of type: ", b1.dtype)
```

**4.3 Matplotlib** - Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python. Matplotlib makes easy things easy and hard things possible.

- Create publication quality plots.
- Make interactive figures that can zoom, pan, update.
- Customize visual style and layout.
- Export to many file formats.
- Embed in JupyterLab and Graphical User Interfaces.
- Use a rich array of third-party packages built on Matplotlib.

üîó Source: Official website of Matplotlib (https://matplotlib.org/)

```python
import matplotlib.pyplot as plt
```

```python
# Simple line plot
x = np.array([1, 2, 3, 4, 5])
y = x**2
plt.plot(x, y)
plt.show()
```

```python
# Simple bar chart
cities = ['New York', 'London', 'Paris']
populations = [8.4, 9.0, 2.1]
plt.bar(cities, populations)
plt.xlabel("City")
plt.ylabel("Population (in millions)")
plt.title("City Populations")
plt.show()
```
